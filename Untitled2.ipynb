```python
import base64
import email
import pickle
import requests
from googleapiclient.discovery import build
from bs4 import BeautifulSoup
import re
import html

# âœ… Load Gmail API credentials
with open("token.pickle", "rb") as token:
    creds = pickle.load(token)

# âœ… Connect to Gmail API
service = build("gmail", "v1", credentials=creds)

def preprocess_email_text(email_text):
    """
    Cleans and preprocesses raw email text.
    - Removes HTML tags
    - Decodes HTML entities
    - Normalizes links
    - Removes extra spaces and special characters
    """
    soup = BeautifulSoup(email_text, "html.parser")  # Remove HTML tags
    text = soup.get_text()
    text = html.unescape(text)  # Decode HTML entities
    text = re.sub(r'http[s]?://\S+', ' [LINK] ', text)  # Normalize links
    text = re.sub(r"(?i)(you received this email|unsubscribe|privacy policy|Â©|all rights reserved).*", "", text)
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

def fetch_and_classify_5_emails():
    """Fetches the last 5 emails, preprocesses them, and classifies them using our API."""
    results = service.users().messages().list(userId="me", maxResults=5).execute()
    messages = results.get("messages", [])

    for i, msg in enumerate(messages, start=1):
        msg_id = msg["id"]
        msg_data = service.users().messages().get(userId="me", id=msg_id, format="raw").execute()
        raw_email = base64.urlsafe_b64decode(msg_data["raw"].encode("ASCII"))
        msg_obj = email.message_from_bytes(raw_email)

        # âœ… Extract subject and email body
        subject = msg_obj["Subject"] if msg_obj["Subject"] else "(No Subject)"
        body = ""

        if msg_obj.is_multipart():
            for part in msg_obj.walk():
                if part.get_content_type() == "text/plain":
                    body = part.get_payload(decode=True).decode(errors="ignore")
        else:
            body = msg_obj.get_payload(decode=True).decode(errors="ignore")

        # âœ… Preprocess email body
        clean_body = preprocess_email_text(body)

        # âœ… Send to Flask API for classification
        api_url = "http://127.0.0.1:5000/classify"  # Change to your Flask API URL
        response = requests.post(api_url, json={"text": clean_body})

        # âœ… Print results
        print(f"\nðŸ“© **Email {i}**")
        print(f"ðŸ”¹ **Subject:** {subject}")
        print(f"\nðŸ“œ **Preprocessed Email Body:**\n{clean_body[:500]}")  # Print first 500 characters

        if response.status_code == 200:
            result = response.json()
            print(f"\nðŸ” **Classification:** {result['label']} (Confidence: {result['confidence']:.2f})")
        else:
            print(f"âš ï¸ Error classifying email: {response.text}")

        print("-" * 80)

# âœ… Fetch, preprocess, and classify 5 emails
fetch_and_classify_5_emails()
```
